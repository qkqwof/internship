```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('data/최종.csv', encoding = 'euc_kr')
df.head(3)
```

```python
features = df.loc[:, ['주차수요량', '카드이용자수', '사업체수']]
features.head(3)
```

# 군집수 판단

```python

kmlist = []
for i in range(1,11):
    km = KMeans(n_clusters=i, algorithm='auto', random_state=42)
    km.fit(features)
    
    kmlist.append(km.inertia_)
plt.plot(range(1,11), kmlist, marker='o')
plt.xticks(range(1,11))
plt.xlabel('number of clusters, k' )
plt.ylabel('inertia')
# 군집 중심에서 각 클러스터 내 점 거리 합산
plt.show()
```

```python

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
minmaxdata= scaler.fit_transform(features)

kmlist = []
for i in range(1,11):
    km = KMeans(n_clusters=i, algorithm='auto', random_state=42)
    km.fit(minmaxdata)
    
    kmlist.append(km.inertia_)
plt.plot(range(1,11), kmlist, marker='o')
plt.xticks(range(1,11))
plt.xlabel('number of clusters, k' )
plt.ylabel('inertia')
# 군집 중심에서 각 클러스터 내 점 거리 합산
plt.show()
```

```python
model = KMeans(n_clusters=3, algorithm='auto')
model.fit(features)

predict = pd.DataFrame(model.predict(features))
predict.columns = ['predict']
predict
```

```python
r = pd.concat([df['읍면동'] ,features, predict], axis = 1)
r
```

```python
plt.scatter(r['주차수요량'], r['카드이용자수'], c = r['predict'], alpha = 0.5)
```

# 스케일링

```python
scaler = StandardScaler()
model = KMeans(n_clusters=3)
pipeline = make_pipeline(scaler, model)
```

```python
pipeline.fit(features)
predict = pd.DataFrame(pipeline.predict(features))
predict.columns = ['predict']
predict
```

```python
a = pd.concat([df['읍면동'] ,features, predict], axis = 1)
a
```

```python
a.to_csv('kmeans.csv', encoding = 'euc_kr')
```

```python
plt.scatter(a['주차수요량'], a['카드이용자수'], c = a['predict'], alpha = 0.5)
plt.show()
```
