```python
from numpy import *
import pandas as np
x = np.array(df["주차수요량"])
x_standard = (x-mean(x, axis = 0))/std(x, axis = 0)
x_standard
```

```python
print('mean : ', mean(x_standard, axis = 0))
print('std : ', std(x_standard, axis = 0) )
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(x_standard, rug = True)
plt.show()
```
