```python
df1 = pd.read_csv('교차로 및 구간별 교통량 정리- 무수천-신광사거리(수정).csv')
df2 = pd.read_csv('교차로 및 구간별 교통량 정리-마리나-거로사거리(수정).csv')
df3 = pd.read_csv('교차로 및 구간별 교통량 정리- 신시가지입구-중앙여고 교차로.csv')
df4 = pd.read_csv('교차로 및 구간별 교통량 정리- 남문사거리-아라초(수정).csv')
```

```python
df11 = df1[df1['latitude_x'].notnull()].reset_index()
df11
```

```python
df21 = df2[df2['latitude_x'].notnull()].reset_index()
df21
```

```python
df31 = df3[df3['latitude_x'].notnull()].reset_index()
df31
```

```python
df41 = df4[df4['latitude_x'].notnull()].reset_index()
df41
```

```python
df_fin = pd.concat([df11,df21,df31,df41],axis=0)
df_fin
```

```python
lines = []
for i in range(len(df_fin)):
    lines.append(df_fin[['latitude_x','longitude_x']].values[i].tolist())
    lines.append(df_fin[['latitude_y','longitude_y']].values[i].tolist())
lines
```

```python
center = [33.37835,126.54598]
m = f.Map(location=center, zoom_start=10)
f.PolyLine(locations=lines,
          tooltip = 'PolyLine').add_to(m)
#m.save('재영.html')
```
